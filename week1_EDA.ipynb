4. Name the Movie which has the highest Runtime ?

```python
max_runtime_index = np.argmax( [ int(str[:-4]) for str in df['Runtime'] ]  )
movie_with_highest_runtime = df.iloc[max_runtime_index].name
movie_with_highest_runtime
```

5. How many movies are released after year 2000 and have IMDB rating > 8.5 ?

```python
count_year_IMDB = 0
for i in range(len(df)):
  if((df.iloc[i]['Released_Year'].isdigit() and int(df.iloc[i]['Released_Year'])>2000) and df.iloc[i]['IMDB_Rating']>8.5):
    count_year_IMDB += 1
print(count_year_IMDB)
```

6. Make a countplot of Movies directed by all directors.

```python
plt.figure(figsize=(200,4))
ax = sns.countplot(x=df['Director'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.xlabel('Director')
plt.ylabel('No.of movies directed for')
plt.title('Count of movies directed by all directors')
plt.show()
```

7. Make a plot to analyze distribution of IMDB rating?

```python
plt.figure(figsize=(10,3))
plt.hist(df['IMDB_Rating'],color = 'orange')
plt.xlabel('IMDB Rating')
plt.ylabel('Frequency')
plt.title('Distribution of IMDB rating')
```
